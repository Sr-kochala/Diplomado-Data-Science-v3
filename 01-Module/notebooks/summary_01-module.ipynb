````python
# Simular datos de temperatura (valores aleatorios entre 15°C y 30°C)
temperaturas = np.random.uniform(15, 30, 168)  # 168 valores

# Calcular estadísticas clave
media = np.mean(temperaturas)
maxima = np.max(temperaturas)
minima = np.min(temperaturas)

# Encontrar horas con temperatura > 28°C
horas_calurosas = np.where(temperaturas > 28)[0]

print(f"Media: {media:.2f}°C")         # Ej: Media: 22.53°C
print(f"Máxima: {maxima:.2f}°C")       # Ej: Máxima: 29.87°C
print(f"Horas calurosas: {horas_calurosas}")  # Ej: [ 12, 45, 120 ]
````

**Probabilidad Binomial**
```
from scipy.stats import binom: 
``` 
importa la distribución binomial del módulo stats de la biblioteca SciPy, permitiendo trabajar con experimentos de probabilidad discreta de tipo "éxito/fracaso"

```
binom.pmf(k, n, p)
```
unción calcula la función de masa de probabilidad (PMF)
¿Cuál es la probabilidad de obtener exactamente k éxitos?

```
binom.cdf(k, n, p)
```
Esta función calcula la función de distribución acumulada  (CDF)
 ¿Cuál es la probabilidad de obtener k o menos éxitos?

Parametros
- **k:** El número de éxitos que nos interesa.
- **n:** El número total de ensayos.
- **p:** La probabilidad de éxito en un solo ensayo.


**Distribución normal**

```
from scipy.stats import norm
```
importa la función norm desde el módulo stats de la biblioteca SciPy. Esta función norm representa la distribución normal, también conocida como la distribución gaussiana

```
norm.pdf(x, loc=0, scale=1)
```
Calcula la función de densidad de probabilidad (PDF), de la distribución normal. 
Nos dice qué tan probable es observar un valor particular en la distribución.

```
norm.ppf(q, loc=0, scale=1)
```
Calcula el punto porcentual (PPF) o cuantil de la distribución normal.
Nos dice qué valor de la variable aleatoria corresponde a un percentil específico.

Parámetros:
- **q:** *La probabilidad acumulada (un valor entre 0 y 1).*
- **loc:** *La media (μ) de la distribución normal. Por defecto, es 0 (distribución normal estándar).*
- **scale:** *La desviación estándar (σ) de la distribución normal. Por defecto, es 1 (distribución normal estándar).*

**Distribucion Poisson**

```
from scipy.stats import poisson as pois
```
La función poisson se utiliza para trabajar con la distribución de Poisson, que es una distribución de probabilidad discreta que describe la probabilidad de un número dado de eventos que ocurren en un intervalo 1  de tiempo o espacio fijo 

```
pois.pmf(k, mu) 
```
Esta función calcula la función de masa de probabilidad (PMF)
Cuál es la probabilidad de que ocurran exactamente k eventos?

```
pois.cdf(k, mu)
```
Esta función calcula la función de distribución acumulada (CDF)
¿Cuál es la probabilidad de que ocurran k o menos eventos?

Parámetros
- **k:** *El número de eventos que nos interesa.*
- **mu:** *La tasa promedio de eventos (también conocida como lambda, λ).*
